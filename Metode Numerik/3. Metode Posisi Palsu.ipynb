### Minta <i>input</i>

```python
e = 2.7182818248590452
a = float(input("Masukan batas bawah : "))
b = float(input("Masukan batas atas  : "))
tolSal = float(input("Masukan toleransi kesalahan : "))
```

### Masukkan nilai perhitungan dari tiap fungsi ke dalam tiap variabel yang telah ditentukan

```python
fa = (a**2)-(e**a)+8         # BISA DIGANTI DENGAN FUNGSI YANG LAIN
fb = (b**2)-(e**b)+8         # BISA DIGANTI DENGAN FUNGSI YANG LAIN
x = b-((fb*(b-a))/(fb-fa))   
fx = (x**2)-(e**x)+8         # BISA DIGANTI DENGAN FUNGSI YANG LAIN
```

### Buat fungsi untuk mencetak nilai

```python
def cetak():
    print("Nilai a ",a)
    print("Nilai x ",x)
    print("Nilai b ",b)
    print("Nilai fa ",fa)
    print("Nilai fx ",fx)
    print("Nilai fb ",fb)
    print("")
cetak()
```

### Lakukan iterasi terus menerus sampai nilai fx atau (fx * -1) lebih kecil dari toleransi kesalahan

```python
while(float(fx) > float(tolSal) or float(fx)*(-1) > float(tolSal)):
    if (fa*fx > 0):
        a=x
        b=b
    else:
        a=a
        b=x
    
    x = b-((fb*(b-a))/(fb-fa))     
    fa = (a**2)-(e**a)+8           # BISA DIGANTI DENGAN FUNGSI YANG LAIN
    fx = (x**2)-(e**x)+8           # BISA DIGANTI DENGAN FUNGSI YANG LAIN
    fb = (b**2)-(e**b)+8           # BISA DIGANTI DENGAN FUNGSI YANG LAIN
    cetak()
```
