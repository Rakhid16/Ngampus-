### Buat fungsi untuk menentukan nilai x yang diminta dengan menggunakan Metode Tabel lalu cetak iterasi pertama

```python
def iterasi():
    jumBag = ((a-b)/10)*(-1)
    
    for i in range(9):
        tabel.append(tabel[i]+jumBag)
        tabel.sort()
        
    for j in range(11):
        tabelBar.append((tabel[j]**2)-(e**tabel[j])+8) # BISA DIGANTI DENGAN FUNGSI YANG LAIN
        
    for k in range(11):
        print(tabel[k])
        print(tabelBar[k])
        if tabelBar[k] > 0:
            x.append(tabelBar[k])
        else:
            y.append(tabelBar[k])
iterasi()
```
