### Minta <i>input</i>

```python
e = 2.7182818248590452
a = float(input("Masukan batas bawah : "))
b = float(input("Masukan batas atas  : "))
```

### Lakukan iterasi terus menerus selagi nilai fx tidak sama dengan 0

```python
while True:
    x = (a+b)/2
    fa = (a**2)-(e**a)+8   # BISA DIGANTI DENGAN FUNGSI YANG LAIN
    fx = (x**2)-(e**x)+8   # BISA DIGANTI DENGAN FUNGSI YANG LAIN
    fb = (b**2)-(e**b)+8   # BISA DIGANTI DENGAN FUNGSI YANG LAIN
    
    print("\nNilai a ",a)
    print("Nilai x ",x)
    print("Nilai b ",b)
    print("Nilai fa ",fa)
    print("Nilai fx ",fx)
    print("Nilai fb ",fb)
    
    if (fa*fx > 0):
        a,b=x,b
    else:
        a,b=a,x
        
    if (fx == 0):
        break
```
